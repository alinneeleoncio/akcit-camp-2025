**1. Variáveis Qualitativas (ou Categóricas)**

São variáveis que não representam quantidade, mas sim categorias, grupos ou atributos.
Elas descrevem qualidades, não números.

**Nominais**

Não possuem ordem natural entre as categorias.
Exemplos:

* Profissão

* Gênero

* Religião

Essas categorias não têm hierarquia — “engenheiro” não é maior nem menor que “médico”.

**Ordinais**

Possuem ordem ou hierarquia, mas sem uma diferença numérica mensurável entre os níveis.
Exemplos:

* Escolaridade (fundamental < médio < superior)

* Estágio da doença (leve < moderado < grave)

* Classe social (baixa < média < alta)

Elas têm ordem, mas não dá para dizer “classe alta é duas vezes classe média”.

**2. Variáveis Quantitativas (ou Numéricas)**

São variáveis que representam quantidades mensuráveis.

**Discretas**

Assumem valores inteiros, normalmente provenientes de contagem.
Exemplos:

* Número de filhos

* Número de acessos a uma plataforma

* Quantidade de produtos

Elas não têm valores fracionários.

**Contínuas**

Podem assumir qualquer valor real dentro de um intervalo.
São medidas, não contagens.
Exemplos:

* Altura

* Peso

* Salário

* Temperatura

Elas podem ter casas decimais.

# **Visualização de Dados**

A visualização de dados é uma etapa essencial da análise, pois permite identificar padrões, comparar categorias, observar tendências e detectar anomalias de maneira intuitiva. A escolha do gráfico correto depende principalmente do tipo de variável (numérica ou categórica) e do objetivo da análise.

Para variáveis numéricas contínuas, gráficos como histogramas e boxplots são ideais: o histograma mostra a distribuição dos valores, enquanto o boxplot destaca dispersão, mediana e possíveis outliers. Quando queremos analisar a relação entre duas variáveis numéricas, o gráfico mais apropriado é o scatter plot (dispersão), que evidencia correlações e padrões conjuntos.

Para variáveis categóricas, utilizamos gráficos que mostram contagem ou proporção, como gráficos de barras ou colunas, adequados para comparar categorias distintas. Quando misturamos um valor numérico e uma categoria (por exemplo, preço médio por fabricante), gráficos de barras continuam sendo a escolha ideal, pois permitem comparar estatísticas agregadas entre grupos.

Em séries temporais — variáveis associadas ao tempo — o gráfico mais adequado é o gráfico de linha, pois mostra a evolução, tendência e sazonalidade ao longo do tempo. Já quando queremos comparar partes de um todo, gráficos de pizza podem ser usados, embora sejam menos precisos e geralmente substituídos por barras horizontais.

```python
import matplotlib.pyplot as plt

plt.rcParams.update({
    "font.family": "serif",
    "font.serif": ["CMU Serif", "Computer Modern", "DejaVu Serif"]
})

cms = 1/2.54
```

```python
plt.figure(figsize=(16*cms,10*cms))
df_limpo["make"].value_counts().plot(kind="bar")
plt.title("Veículos x marca")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16*cms,10*cms))
df_limpo["price"].hist(edgecolor="black")
plt.title("Distribuição de preços")
plt.ylabel("Frequência/Contagem")
plt.xlabel("Unidade Monetário (U$)")
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16*cms,10*cms))
plt.boxplot(df_limpo["city_kmpl"])
plt.title("Análise de outliers para km/L na cidade")
plt.ylabel("km/L na cidade")
plt.tight_layout()
plt.show()
```

```python
df_limpo.loc[df_limpo["city_kmpl"]>16]["make"]
```

```python
## groupby()

kml_por_marca = df_limpo.groupby("make")["city_kmpl"].mean().reset_index()

kml_por_marca
```

```python
plt.figure(figsize=(16*cms,10*cms))
plt.bar(kml_por_marca["make"],kml_por_marca["city_kmpl"],edgecolor="black")
plt.title("km/L na cidade por fabricante")
plt.ylabel("km/L")
plt.xticks(rotation=90)

plt.hlines(df_limpo["city_kmpl"].mean(),xmin=0,xmax=len(kml_por_marca["make"])-1,color="red",linestyles="dashed")

plt.tight_layout()
plt.show()
```
